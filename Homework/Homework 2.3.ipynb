### Homework 2.3 - 2D Temperature Distribution

Solve 2D heat equation for the following initial conditions:

$$\begin{array}{l}
u(x,0) = \sin (2\pi x)\\
u(x,1) = \sin (2\pi x)\\
u(0,y) = \sin (3\pi y)\\
u(1,y) = \sin (\pi y)
\end{array}$$

where $u(x, y)$ represent the steady state temperature distribution on 2D plane.
 
Plot the 2D heat map of the solution.

```python
n = 100
maxIter = 500
U = np.empty(shape=(n, n))
U.fill(0.)
for i in range(n):
    U[i, 0] = math.sin(2*math.pi*i/n)
    U[i, n-1] = math.sin(2*math.pi*i/n)
for j in range(n):
    U[0, j] = math.sin(3*math.pi*j/n)
    U[n-1, j] = math.sin(math.pi*j/n)

for _ in range(maxIter):
    Uold = np.copy(U)
    for i in range(1,n-1):
        for j in range(1,n-1):
            U[i, j] = 0.25 * (Uold[i+1, j] + Uold[i-1, j] + Uold[i, j+1] + Uold[i, j-1])

plt.imshow(U, cmap="hot")
plt.colorbar()

plt.show()
```
